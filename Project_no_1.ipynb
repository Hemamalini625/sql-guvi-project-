 ***LEFT OUTER JOIN:***

 In left outer join of 2 data frames,the first (left) data frame will print as it is.
 And the second data frame the values which is common in first data frame only printed.

# **We use MERGE method and how ='left'**

 ***RIGHT OUTER JOIN:***


 In Right outer join of 2 data frames,the Second (right) data frame will print as it is.
 And the first data frame the values which is common in second data frame only printed.

# **We use MERGE method and how ='right'**

**INNER JOIN:**

In inner join function, with the 2 data frames the common values in both data frames only printed

# **We use MERGE method and how ='inner'**


**OUTER JOIN:**

In 2 data frames , the outer join print all the values in both data frames and fill NAN values in which there is no common values.

# **We use MERGE method and how ='outer'**
